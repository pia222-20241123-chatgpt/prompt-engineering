```python
import requests
from bs4 import BeautifulSoup

# 네이버 뉴스 URL
url = 'https://news.naver.com/'

# HTTP 요청을 보내서 페이지를 가져옵니다.
response = requests.get(url)

# 요청이 성공하면 HTML 페이지를 파싱합니다.
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # 'class'가 'cnf_news_title'인 <strong> 태그를 찾습니다.
    headlines = soup.find_all('strong', {'class': 'cnf_news_title'})

    if headlines:
        for idx, headline in enumerate(headlines):
            print(f'{idx+1}. {headline.get_text().strip()}')
    else:
        print("헤드라인 뉴스가 없습니다.")
else:
    print('페이지를 가져오지 못했습니다.')
```

```python
import requests
from bs4 import BeautifulSoup

# Daum 뉴스 URL
url = "https://news.daum.net/"

# HTTP 요청을 보내고 HTML 응답 받기
response = requests.get(url)

# 응답 인코딩을 'utf-8'로 설정
response.encoding = 'utf-8'

# HTML 내용 가져오기
html_content = response.text

# BeautifulSoup을 사용해 HTML 파싱
soup = BeautifulSoup(html_content, "html.parser")

# 'strong' 태그에서 클래스명이 'tit_txt'인 요소 찾기
news_titles = soup.find_all("strong", class_="tit_txt")

# 뉴스 제목 출력
for idx, title in enumerate(news_titles, start=1):
    print(f"{idx}. {title.get_text(strip=True)}")
```
