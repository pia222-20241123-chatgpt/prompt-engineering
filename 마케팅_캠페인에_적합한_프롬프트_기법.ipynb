1. 목표 설정 : 캠페인의 목적(인지도 향상, 전환율 증가, 브랜딩 등)을 명확히 정의
2. 타겟 설정: 특정 고객층(성별, 연령, 관심사 등)을 정확히 설정
3. CTA(Call to Action) : 사용자 행동을 유도하는 강력한 메세지
4. 스토리 텔링 : 감성적으로 공감할 수 있는 스토리를 활용
5. 문제 해결형 프롬프트 : 고객의 문제를 정의하고 이를 해결하는 방법을 제안
6. 짧고 명확한 문구 : 간결하면서도 메세지가 강렬한 표현을 사용
7. 소셜 증명 포함 : 고객 리뷰나 사례를 언급
8. 긴급성 강조 : 제한된 시간, 할인등의 요소
9. 유모와 독창성 : 주목받을 수 있는 창의적인 요소를 추가
10. 채널 맞춤형 톤: :사용하는 플렛폼에 다라 적합한 톤과 스타일을 설정
