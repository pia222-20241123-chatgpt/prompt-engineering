```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV 파일 읽기
file_path = '/content/주택도시보증공사_전국 신규 민간아파트 분양가격 동향_20240731.csv'  # 실제 CSV 파일 경로로 변경하세요.
df = pd.read_csv(file_path,encoding='cp949')

# 데이터프레임의 처음 몇 줄 출력
print("데이터프레임의 처음 몇 줄:")
print(df.head())

# 데이터프레임 기본 정보 출력
print("\n데이터프레임 기본 정보:")
print(df.info())

# 데이터프레임 요약 통계 출력
print("\n데이터프레임 요약 통계:")
print(df.describe())

# 결측치 확인
print("\n데이터프레임의 결측치:")
print(df.isnull().sum())

# 분양가격(제곱미터)의 분포
plt.figure(figsize=(10, 6))
sns.histplot(df['분양가격(제곱미터)'], kde=True)
plt.title('분양가격(제곱미터) 분포')
plt.xlabel('분양가격(제곱미터)')
plt.ylabel('빈도')
plt.show()

# 연도별 평균 분양가격
plt.figure(figsize=(12, 6))
sns.barplot(x='연도', y='분양가격(제곱미터)', data=df)
plt.title('연도별 평균 분양가격')
plt.xlabel('연도')
plt.ylabel('평균 분양가격(제곱미터)')
plt.show()

# 지역별 평균 분양가격
plt.figure(figsize=(14, 8))
sns.barplot(x='분양가격(제곱미터)', y='지역명', data=df, ci=None)
plt.title('지역별 평균 분양가격')
plt.xlabel('평균 분양가격(제곱미터)')
plt.ylabel('지역명')
plt.show()

# 상관관계 히트맵
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('상관관계 히트맵')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['axes.unicode_minus'] = False  # 음수 기호 깨짐 방지

# 1. 데이터 읽기
file_path = '/content/주택도시보증공사_전국 신규 민간아파트 분양가격 동향_20240731.csv'  # 파일 경로를 실제 경로로 수정하세요.
data = pd.read_csv(file_path,encoding='cp949')

# 2. 컬럼명 변경
data.columns = ['Region', 'Size', 'Year', 'Month', 'Price_per_sq_meter']

# 3. 분양가격(제곱미터) 컬럼을 숫자형으로 변환 (문자열로 되어있는 경우)
data['Price_per_sq_meter'] = pd.to_numeric(data['Price_per_sq_meter'], errors='coerce')

# 4. 결측치 처리: 분양가격이 없는 행은 제거
data = data.dropna(subset=['Price_per_sq_meter'])

# 5. 데이터 정보 확인
print("First 5 rows of the data:")
print(data.head())

print("\nData Info:")
print(data.info())

print("\nData Description:")
print(data.describe())

# 6. Price distribution by region
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y='Price_per_sq_meter', data=data)
plt.xticks(rotation=45)
plt.title('Price Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Price per Square Meter')
plt.tight_layout()
plt.show()

# 7. Price distribution by size category
plt.figure(figsize=(12, 6))
sns.boxplot(x='Size', y='Price_per_sq_meter', data=data)
plt.title('Price Distribution by Size Category')
plt.xlabel('Size')
plt.ylabel('Price per Square Meter')
plt.tight_layout()
plt.show()

# 8. Price trend over years
plt.figure(figsize=(14, 6))
sns.lineplot(x='Year', y='Price_per_sq_meter', data=data)
plt.title('Price Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Price per Square Meter')
plt.tight_layout()
plt.show()

# 9. Average price per month
monthly_avg_price = data.groupby('Month')['Price_per_sq_meter'].mean()
plt.figure(figsize=(14, 6))
monthly_avg_price.plot(kind='bar', color='skyblue')
plt.title('Average Price per Square Meter by Month')
plt.xlabel('Month')
plt.ylabel('Average Price per Square Meter')
plt.tight_layout()
plt.show()

# 10. Average price by region
regional_avg_price = data.groupby('Region')['Price_per_sq_meter'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
regional_avg_price.plot(kind='bar', color='lightcoral')
plt.title('Average Price per Square Meter by Region')
plt.xlabel('Region')
plt.ylabel('Average Price per Square Meter')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 11. Price distribution (Histogram)
plt.figure(figsize=(12, 6))
sns.histplot(data['Price_per_sq_meter'], bins=30, kde=True)
plt.title('Distribution of Price per Square Meter')
plt.xlabel('Price per Square Meter')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 12. Heatmap of average price by year and month
plt.figure(figsize=(14, 8))
sns.heatmap(data.pivot_table(index='Year', columns='Month', values='Price_per_sq_meter', aggfunc='mean'),
            cmap='YlGnBu', annot=True, fmt=".1f", linewidths=0.5)
plt.title('Average Price per Square Meter by Year and Month')
plt.xlabel('Month')
plt.ylabel('Year')
plt.tight_layout()
plt.show()
```
