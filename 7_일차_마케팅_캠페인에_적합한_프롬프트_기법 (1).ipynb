```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú
file_path = '/content/Ï£ºÌÉùÎèÑÏãúÎ≥¥Ï¶ùÍ≥µÏÇ¨_Ï†ÑÍµ≠ Ïã†Í∑ú ÎØºÍ∞ÑÏïÑÌååÌä∏ Î∂ÑÏñëÍ∞ÄÍ≤© ÎèôÌñ•_20240731.csv'  # ÌååÏùº Í≤ΩÎ°úÎ•º ÏûÖÎ†•ÌïòÏÑ∏Ïöî.
data = pd.read_csv(file_path, encoding='euc-kr')

# 'Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)' Ïª¨ÎüºÏùÑ Ïà´ÏûêÎ°ú Î≥ÄÌôò (Í≤∞Ï∏°Í∞í Ï≤òÎ¶¨ Ìè¨Ìï®)
data['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'] = pd.to_numeric(data['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'], errors='coerce')

# Í≤∞Ï∏°Í∞í Ï†úÍ±∞
data_cleaned = data.dropna(subset=['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'])

# Ïó∞ÎèÑÎ≥Ñ, ÏßÄÏó≠Î≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä Î∂ÑÏÑù
avg_price_by_year_region = data_cleaned.groupby(['Ïó∞ÎèÑ', 'ÏßÄÏó≠Î™Ö'])['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'].mean().unstack()

# ÏöîÏïΩ
summary = avg_price_by_year_region.describe()
print("Ïó∞ÎèÑÎ≥Ñ ÏßÄÏó≠Î≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä ÏöîÏïΩ:")
print(summary)

# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12, 8))
sns.heatmap(avg_price_by_year_region, annot=True, fmt=".0f", cmap="coolwarm", linewidths=.5)
plt.title("Ïó∞ÎèÑÎ≥Ñ ÏßÄÏó≠Î≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä", fontsize=16)
plt.xlabel("ÏßÄÏó≠Î™Ö", fontsize=12)
plt.ylabel("Ïó∞ÎèÑ", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Í≤∞Í≥º Ï†ÄÏû• (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
plt.savefig('Ïó∞ÎèÑÎ≥Ñ_ÏßÄÏó≠Î≥Ñ_ÌèâÍ∑†_Î∂ÑÏñëÍ∞Ä_ÌûàÌä∏Îßµ.png')
plt.show()

# ÎùºÏù∏ ÌîåÎ°Ø: Ïó∞ÎèÑÎ≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä Î≥ÄÌôî
plt.figure(figsize=(12, 8))
avg_price_by_year = data_cleaned.groupby('Ïó∞ÎèÑ')['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'].mean()
avg_price_by_year.plot(marker='o')
plt.title("Ïó∞ÎèÑÎ≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä Ï∂îÏù¥", fontsize=16)
plt.xlabel("Ïó∞ÎèÑ", fontsize=12)
plt.ylabel("ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä (Ï†úÍ≥±ÎØ∏ÌÑ∞Îãπ)", fontsize=12)
plt.grid(True)
plt.tight_layout()

# Í≤∞Í≥º Ï†ÄÏû• (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
plt.savefig('Ïó∞ÎèÑÎ≥Ñ_ÌèâÍ∑†_Î∂ÑÏñëÍ∞Ä_Ï∂îÏù¥.png')
plt.show()

# Í≤∞Í≥º ÏöîÏïΩ Ï†ÄÏû• (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
summary.to_csv('Ïó∞ÎèÑÎ≥Ñ_ÏßÄÏó≠Î≥Ñ_Î∂ÑÏñëÍ∞Ä_ÏöîÏïΩ.csv', encoding='euc-kr')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
import numpy as np

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú
file_path = '/content/Ï£ºÌÉùÎèÑÏãúÎ≥¥Ï¶ùÍ≥µÏÇ¨_Ï†ÑÍµ≠ Ïã†Í∑ú ÎØºÍ∞ÑÏïÑÌååÌä∏ Î∂ÑÏñëÍ∞ÄÍ≤© ÎèôÌñ•_20240731.csv'  # ÌååÏùº Í≤ΩÎ°úÎ•º ÏûÖÎ†•ÌïòÏÑ∏Ïöî.
data = pd.read_csv(file_path, encoding='euc-kr')

# 'Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)' Ïª¨ÎüºÏùÑ Ïà´ÏûêÎ°ú Î≥ÄÌôò (Í≤∞Ï∏°Í∞í Ï≤òÎ¶¨ Ìè¨Ìï®)
data['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'] = pd.to_numeric(data['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'], errors='coerce')

# Í≤∞Ï∏°Í∞í Ï†úÍ±∞
data_cleaned = data.dropna(subset=['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'])

# Ïó∞ÎèÑÎ≥Ñ, ÏßÄÏó≠Î≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä Î∂ÑÏÑù
avg_price_by_year_region = data_cleaned.groupby(['Ïó∞ÎèÑ', 'ÏßÄÏó≠Î™Ö'])['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'].mean().unstack()

# ÏöîÏïΩ
summary = avg_price_by_year_region.describe()
print("Ïó∞ÎèÑÎ≥Ñ ÏßÄÏó≠Î≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä ÏöîÏïΩ:")
print(summary)

# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12, 8))
sns.heatmap(avg_price_by_year_region, annot=True, fmt=".0f", cmap="coolwarm", linewidths=.5)
plt.title("Ïó∞ÎèÑÎ≥Ñ ÏßÄÏó≠Î≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä", fontsize=16)
plt.xlabel("ÏßÄÏó≠Î™Ö", fontsize=12)
plt.ylabel("Ïó∞ÎèÑ", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Í≤∞Í≥º Ï†ÄÏû• (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
plt.savefig('Ïó∞ÎèÑÎ≥Ñ_ÏßÄÏó≠Î≥Ñ_ÌèâÍ∑†_Î∂ÑÏñëÍ∞Ä_ÌûàÌä∏Îßµ.png')
plt.show()

# ÎùºÏù∏ ÌîåÎ°Ø: Ïó∞ÎèÑÎ≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä Î≥ÄÌôî
plt.figure(figsize=(12, 8))
avg_price_by_year = data_cleaned.groupby('Ïó∞ÎèÑ')['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'].mean()
avg_price_by_year.plot(marker='o')
plt.title("Ïó∞ÎèÑÎ≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä Ï∂îÏù¥", fontsize=16)
plt.xlabel("Ïó∞ÎèÑ", fontsize=12)
plt.ylabel("ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä (Ï†úÍ≥±ÎØ∏ÌÑ∞Îãπ)", fontsize=12)
plt.grid(True)
plt.tight_layout()

# Í≤∞Í≥º Ï†ÄÏû• (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
plt.savefig('Ïó∞ÎèÑÎ≥Ñ_ÌèâÍ∑†_Î∂ÑÏñëÍ∞Ä_Ï∂îÏù¥.png')
plt.show()

# Î©¥Ï†ÅÎ≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä Î∂ÑÏÑù Î∞è ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12, 8))
avg_price_by_size = data_cleaned.groupby('Í∑úÎ™®Íµ¨Î∂Ñ')['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'].mean().sort_values()
avg_price_by_size.plot(kind='barh', color='skyblue')
plt.title("Î©¥Ï†ÅÎ≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä", fontsize=16)
plt.xlabel("ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä (Ï†úÍ≥±ÎØ∏ÌÑ∞Îãπ)", fontsize=12)
plt.ylabel("Î©¥Ï†Å Íµ¨Î∂Ñ", fontsize=12)
plt.tight_layout()

# Í≤∞Í≥º Ï†ÄÏû• (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
plt.savefig('Î©¥Ï†ÅÎ≥Ñ_ÌèâÍ∑†_Î∂ÑÏñëÍ∞Ä_ÎßâÎåÄÍ∑∏ÎûòÌîÑ.png')
plt.show()

# Í≤∞Í≥º ÏöîÏïΩ Ï†ÄÏû• (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
summary.to_csv('Ïó∞ÎèÑÎ≥Ñ_ÏßÄÏó≠Î≥Ñ_Î∂ÑÏñëÍ∞Ä_ÏöîÏïΩ.csv', encoding='euc-kr')

# 2025ÎÖÑ Î∂ÑÏñëÍ∞Ä ÏòàÏ∏° (ÎπÑÏÑ†Ìòï Î™®Îç∏)
# Ïó∞ÎèÑÎ≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
avg_price_by_year = data_cleaned.groupby('Ïó∞ÎèÑ')['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'].mean().reset_index()
X = avg_price_by_year['Ïó∞ÎèÑ'].values.reshape(-1, 1)
y = avg_price_by_year['Î∂ÑÏñëÍ∞ÄÍ≤©(Ï†úÍ≥±ÎØ∏ÌÑ∞)'].values

# Îã§Ìï≠ ÌöåÍ∑Ä Î™®Îç∏ ÌïôÏäµ
poly_model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
poly_model.fit(X, y)

# 2025ÎÖÑ ÏòàÏ∏°
future_year = np.array([[2025]])
future_price = poly_model.predict(future_year)
print(f"2025ÎÖÑ ÏòàÏÉÅ Î∂ÑÏñëÍ∞ÄÍ≤© (ÎπÑÏÑ†Ìòï): {future_price[0]:.2f} (Ï†úÍ≥±ÎØ∏ÌÑ∞Îãπ)")

# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12, 8))
plt.scatter(X, y, color='blue', label='Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞')
plt.plot(np.linspace(X.min(), X.max(), 100).reshape(-1, 1),
         poly_model.predict(np.linspace(X.min(), X.max(), 100).reshape(-1, 1)),
         color='red', label='ÎπÑÏÑ†Ìòï Ï∂îÏÑ∏ÏÑ†')
plt.scatter(future_year, future_price, color='green', label='2025ÎÖÑ ÏòàÏ∏° Í∞í', s=100)
plt.title("Ïó∞ÎèÑÎ≥Ñ ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä Î∞è 2025ÎÖÑ ÏòàÏ∏° (ÎπÑÏÑ†Ìòï)", fontsize=16)
plt.xlabel("Ïó∞ÎèÑ", fontsize=12)
plt.ylabel("ÌèâÍ∑† Î∂ÑÏñëÍ∞Ä (Ï†úÍ≥±ÎØ∏ÌÑ∞Îãπ)", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Í≤∞Í≥º Ï†ÄÏû• (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
plt.savefig('2025ÎÖÑ_Î∂ÑÏñëÍ∞Ä_ÏòàÏ∏°_ÎπÑÏÑ†Ìòï.png')
plt.show()
```
