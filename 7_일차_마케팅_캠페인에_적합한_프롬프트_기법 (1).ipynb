```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드
file_path = '/content/주택도시보증공사_전국 신규 민간아파트 분양가격 동향_20240731.csv'  # 파일 경로를 입력하세요.
data = pd.read_csv(file_path, encoding='euc-kr')

# '분양가격(제곱미터)' 컬럼을 숫자로 변환 (결측값 처리 포함)
data['분양가격(제곱미터)'] = pd.to_numeric(data['분양가격(제곱미터)'], errors='coerce')

# 결측값 제거
data_cleaned = data.dropna(subset=['분양가격(제곱미터)'])

# 연도별, 지역별 평균 분양가 분석
avg_price_by_year_region = data_cleaned.groupby(['연도', '지역명'])['분양가격(제곱미터)'].mean().unstack()

# 요약
summary = avg_price_by_year_region.describe()
print("연도별 지역별 평균 분양가 요약:")
print(summary)

# 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(avg_price_by_year_region, annot=True, fmt=".0f", cmap="coolwarm", linewidths=.5)
plt.title("연도별 지역별 평균 분양가", fontsize=16)
plt.xlabel("지역명", fontsize=12)
plt.ylabel("연도", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# 결과 저장 (선택 사항)
plt.savefig('연도별_지역별_평균_분양가_히트맵.png')
plt.show()

# 라인 플롯: 연도별 평균 분양가 변화
plt.figure(figsize=(12, 8))
avg_price_by_year = data_cleaned.groupby('연도')['분양가격(제곱미터)'].mean()
avg_price_by_year.plot(marker='o')
plt.title("연도별 평균 분양가 추이", fontsize=16)
plt.xlabel("연도", fontsize=12)
plt.ylabel("평균 분양가 (제곱미터당)", fontsize=12)
plt.grid(True)
plt.tight_layout()

# 결과 저장 (선택 사항)
plt.savefig('연도별_평균_분양가_추이.png')
plt.show()

# 결과 요약 저장 (선택 사항)
summary.to_csv('연도별_지역별_분양가_요약.csv', encoding='euc-kr')
```

```python
# 면적별 평균 분양가 분석 및 시각화
plt.figure(figsize=(12, 8))
avg_price_by_size = data_cleaned.groupby('규모구분')['분양가격(제곱미터)'].mean().sort_values()
avg_price_by_size.plot(kind='barh', color='skyblue')
plt.title("면적별 평균 분양가", fontsize=16)
plt.xlabel("평균 분양가 (제곱미터당)", fontsize=12)
plt.ylabel("면적 구분", fontsize=12)
plt.tight_layout()

# 결과 저장 (선택 사항)
plt.savefig('면적별_평균_분양가_막대그래프.png')
plt.show()

# 결과 요약 저장 (선택 사항)
summary.to_csv('연도별_지역별_분양가_요약.csv', encoding='euc-kr')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
import numpy as np

# 데이터 로드
file_path = '/content/주택도시보증공사_전국 신규 민간아파트 분양가격 동향_20240731.csv'  # 파일 경로를 입력하세요.
data = pd.read_csv(file_path, encoding='euc-kr')

# '분양가격(제곱미터)' 컬럼을 숫자로 변환 (결측값 처리 포함)
data['분양가격(제곱미터)'] = pd.to_numeric(data['분양가격(제곱미터)'], errors='coerce')

# 결측값 제거
data_cleaned = data.dropna(subset=['분양가격(제곱미터)'])

# 연도별, 지역별 평균 분양가 분석
avg_price_by_year_region = data_cleaned.groupby(['연도', '지역명'])['분양가격(제곱미터)'].mean().unstack()

# 요약
summary = avg_price_by_year_region.describe()
print("연도별 지역별 평균 분양가 요약:")
print(summary)

# 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(avg_price_by_year_region, annot=True, fmt=".0f", cmap="coolwarm", linewidths=.5)
plt.title("연도별 지역별 평균 분양가", fontsize=16)
plt.xlabel("지역명", fontsize=12)
plt.ylabel("연도", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# 결과 저장 (선택 사항)
plt.savefig('연도별_지역별_평균_분양가_히트맵.png')
plt.show()

# 라인 플롯: 연도별 평균 분양가 변화
plt.figure(figsize=(12, 8))
avg_price_by_year = data_cleaned.groupby('연도')['분양가격(제곱미터)'].mean()
avg_price_by_year.plot(marker='o')
plt.title("연도별 평균 분양가 추이", fontsize=16)
plt.xlabel("연도", fontsize=12)
plt.ylabel("평균 분양가 (제곱미터당)", fontsize=12)
plt.grid(True)
plt.tight_layout()

# 결과 저장 (선택 사항)
plt.savefig('연도별_평균_분양가_추이.png')
plt.show()

# 면적별 평균 분양가 분석 및 시각화
plt.figure(figsize=(12, 8))
avg_price_by_size = data_cleaned.groupby('규모구분')['분양가격(제곱미터)'].mean().sort_values()
avg_price_by_size.plot(kind='barh', color='skyblue')
plt.title("면적별 평균 분양가", fontsize=16)
plt.xlabel("평균 분양가 (제곱미터당)", fontsize=12)
plt.ylabel("면적 구분", fontsize=12)
plt.tight_layout()

# 결과 저장 (선택 사항)
plt.savefig('면적별_평균_분양가_막대그래프.png')
plt.show()

# 결과 요약 저장 (선택 사항)
summary.to_csv('연도별_지역별_분양가_요약.csv', encoding='euc-kr')

# 2025년 분양가 예측 (비선형 모델)
# 연도별 평균 분양가 데이터 준비
avg_price_by_year = data_cleaned.groupby('연도')['분양가격(제곱미터)'].mean().reset_index()
X = avg_price_by_year['연도'].values.reshape(-1, 1)
y = avg_price_by_year['분양가격(제곱미터)'].values

# 다항 회귀 모델 학습
poly_model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
poly_model.fit(X, y)

# 2025년 예측
future_year = np.array([[2025]])
future_price = poly_model.predict(future_year)
print(f"2025년 예상 분양가격 (비선형): {future_price[0]:.2f} (제곱미터당)")

# 시각화
plt.figure(figsize=(12, 8))
plt.scatter(X, y, color='blue', label='실제 데이터')
plt.plot(np.linspace(X.min(), X.max(), 100).reshape(-1, 1),
         poly_model.predict(np.linspace(X.min(), X.max(), 100).reshape(-1, 1)),
         color='red', label='비선형 추세선')
plt.scatter(future_year, future_price, color='green', label='2025년 예측 값', s=100)
plt.title("연도별 평균 분양가 및 2025년 예측 (비선형)", fontsize=16)
plt.xlabel("연도", fontsize=12)
plt.ylabel("평균 분양가 (제곱미터당)", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# 결과 저장 (선택 사항)
plt.savefig('2025년_분양가_예측_비선형.png')
plt.show()
```
