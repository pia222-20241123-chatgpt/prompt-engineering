```python
import requests
import json
import pandas as pd

query = input('나라장터 검색어 입력 : ')

# 입찰공고 정보
url = 'http://apis.data.go.kr/1230000/PubDataOpnStdService/getDataSetOpnStdBidPblancInfo'
total = []

for i in range(1, 100):  # 1부터 99까지 페이지를 요청
  try:
    params = {
        'serviceKey': '4dr0Ui4/QMen1/waNp8qkB24tz549lFOnJIJl/BzZami5c434Bui7bVOp5j1EVKBJzQF2VCgW8QY2lbbXyhqNA==',
        'pageNo': f'{i}',  # 페이지 번호
        'numOfRows': '100',  # 한 페이지에 100개 항목
        'type': 'json',
        'bidNtceBgnDt': '202412060000',
        'bidNtceEndDt': '202412062359'
    }

    response = requests.get(url, params=params)
    response_data = response.content.decode('utf-8')
    json_data = json.loads(response_data)

    # 'items'에 입찰 공고 정보가 있는 경우
    if 'items' in json_data['response']['body']:
        for item in json_data['response']['body']['items']:
            bid_ntce_nm = item.get('bidNtceNm', '')  # 입찰 공고 제목
            bid_ntce_url = item.get('bidNtceUrl', '')  # 입찰 공고 URL

            # 제목 입력한 검색어가 포함되었는지 확인
            if query in bid_ntce_nm:
                total.append({'title': bid_ntce_nm, 'url': bid_ntce_url})
  except:
    pass

# # 결과 출력
df = pd.DataFrame(total)
# df.to_csv('result.csv', index=False, encoding='utf-8')
# df['url']


import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
import pandas as pd


# 이메일 제목과 본문 설정
subject = "입찰 공고 링크"
body = """
    <html>
    <body>
        <p>안녕하세요,</p>
        <p>다음의 입찰 공고 링크를 확인해 주세요:</p>
        <ul>
"""

# df['url']을 사용하여 동적으로 링크 목록을 생성
for idx, url in enumerate(df['url']):
    body += f'<li><a href="{url}">입찰 공고 {idx + 1}</a></li>'

# 이메일 본문 끝부분
body += """
        </ul>
        <p>감사합니다.</p>
    </body>
    </html>
"""

# HTML 본문을 파일로 저장
html_filename = "bid_links.html"
with open(html_filename, "w", encoding="utf-8") as f:
    f.write(body)

# 이메일 보내기
# Gmail 계정 정보
sender_email = 'pia222.20241123.chatgpt@gmail.com'  # 보낼 이메일 주소
receiver_email = 'pia222.20241123.chatgpt@gmail.com'  # 받는 이메일 주소
password = 'mcnt rabo tpyj wdit'  # 비밀번호 (Gmail 계정의 앱 비밀번호 사용 권장)


msg = MIMEMultipart()
msg['From'] = sender_email
msg['To'] = receiver_email
msg['Subject'] = subject

# 이메일 본문 추가
msg.attach(MIMEText("본문은 HTML 첨부파일로 보내드립니다.", 'plain'))

# HTML 파일 첨부
with open(html_filename, "rb") as attachment:
    part = MIMEBase('application', 'octet-stream')
    part.set_payload(attachment.read())
    encoders.encode_base64(part)
    part.add_header('Content-Disposition', f'attachment; filename={html_filename}')
    msg.attach(part)

# SMTP 서버에 연결하여 이메일 전송
try:
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()  # 보안 연결 시작
    server.login(sender_email, password)  # 로그인
    text = msg.as_string()
    server.sendmail(sender_email, receiver_email, text)  # 이메일 전송
    server.quit()  # 서버 종료
    print("이메일 전송 완료!")
except Exception as e:
    print(f"이메일 전송 실패: {e}")
```
